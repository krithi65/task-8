```python
import pandas as pd
df = pd.read_csv("English.csv")
df.head()
```

```python
eng_words = []
for i in range(175621):
    eng_words.append(df.loc[:,"English words/sentences"][i])
eng_words
```

translate english to french

```python
from translate import Translator
french_trans= Translator(to_lang="French")#display 100 words in french
T = []
for i in range(100):
    T.append(french_trans.translate(eng_words[i]))
print(T)
```

translate english to spanish

```python
from translate import Translator#display 100 words in spanish
spanish_trans= Translator(to_lang="spanish")
T = []
for i in range(100):
    T.append(spanish_trans.translate(eng_words[i]))
print(T)
```

translate english to german

Task-2
	Create a program that needs to automatically correct that spelling from the word or a given sentence. (Language : English) 
Note:  show 100 data point outputs both in words and sentence. 


```python
# importing the nltk suite 
import nltk
  
# importing jaccard distance
# and ngrams from nltk.util
from nltk.metrics.distance import jaccard_distance
from nltk.util import ngrams
```

```python
# Downloading and importing
# package 'words' from nltk corpus
nltk.download('words')
from nltk.corpus import words
  
  
correct_words = words.words()
```

```python
incorrect_words=["cmputr", "watr", "wriet", "athist", "acceptible", "benifit", "acommodate", "amatuer", "acknowlege", "becomeing", 
                            "camoflague", "catagory", "cauhgt", "cematery", "apparrent", "congradulate", "conceed", "definit", "disasterous", "dilema",
                            "equiptment ", "excede", "experiance", "innoculate", "extreem", "guage", "gratful", "guidence", "heirarchy", "dumbell",
                            "necessery", "inteligence", "kernal", "libary", "outragous", "marshmellow", "miniture", "medeval", "nieghbor", "origenal"]
```

```python
for word in incorrect_words:
    temp = [(jaccard_distance(set(ngrams(word, 2)),
                              set(ngrams(w, 2))),w)
            for w in correct_words if w[0]==word[0]]
    print(sorted(temp, key = lambda val:val[0])[0][1])
```
